    - Basis: ROUGE scores - mid values

### Observations:

- The fine-tuning process significantly improved the transformer's performance across all ROUGE metrics:

    - ROUGE-1 and ROUGE-LSUM show over 100% improvement in F1-score.

- These results indicate that fine-tuning has greatly enhanced the model's ability to generate accurate and relevant summaries. 

- Hence, the fine-tuning of the transformer model is successful. 
 


```python
#MID ROUGE scores

before_finetuning = {
    'rouge1': {
        'precision': 0.30968324775739353,
        'recall': 0.4396760585008399,
        'fmeasure': 0.29054592962775116
    },
    'rouge2': {
        'precision': 0.12535111393152845,
        'recall': 0.17781582343023616,
        'fmeasure': 0.11609005370534525
    },
    'rougeL': {
        'precision': 0.20093013240942154,
        'recall': 0.2907883810753654,
        'fmeasure': 0.1878558128893604
    },
    'rougeLsum': {
        'precision': 0.24092968768212347,
        'recall': 0.34497937380932003,
        'fmeasure': 0.22708400391795613
    }
}


after_finetuning = {
    'rouge1': {
        'precision': 0.6592064440736602,
        'recall': 0.6324733143712006,
        'fmeasure': 0.6132238861334389
    },
    'rouge2': {
        'precision': 0.5080411090198691,
        'recall': 0.4968806580602876,
        'fmeasure': 0.47866874404111104
    },
    'rougeL': {
        'precision': 0.5735140296805632,
        'recall': 0.560047355602015,
        'fmeasure': 0.5396991258438945
    },
    'rougeLsum': {
        'precision': 0.6167078297893005,
        'recall': 0.5946188830525441,
        'fmeasure': 0.5762430613700162
    }
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dictionary with the evaluation metrics and their corresponding ROUGE scores before and after fine-tuning
data = {
    'Metric': ['Precision', 'Recall', 'F1-score'],
    'ROUGE-1 Before': [before_finetuning['rouge1']['precision'], before_finetuning['rouge1']['recall'], before_finetuning['rouge1']['fmeasure']],
    'ROUGE-1 After': [after_finetuning['rouge1']['precision'], after_finetuning['rouge1']['recall'], after_finetuning['rouge1']['fmeasure']],
    'ROUGE-2 Before': [before_finetuning['rouge2']['precision'], before_finetuning['rouge2']['recall'], before_finetuning['rouge2']['fmeasure']],
    'ROUGE-2 After': [after_finetuning['rouge2']['precision'], after_finetuning['rouge2']['recall'], after_finetuning['rouge2']['fmeasure']],
    'ROUGE-L Before': [before_finetuning['rougeL']['precision'], before_finetuning['rougeL']['recall'], before_finetuning['rougeL']['fmeasure']],
    'ROUGE-L After': [after_finetuning['rougeL']['precision'], after_finetuning['rougeL']['recall'], after_finetuning['rougeL']['fmeasure']],
    'ROUGE-LSUM Before': [before_finetuning['rougeLsum']['precision'], before_finetuning['rougeLsum']['recall'], before_finetuning['rougeLsum']['fmeasure']],
    'ROUGE-LSUM After': [after_finetuning['rougeLsum']['precision'], after_finetuning['rougeLsum']['recall'], after_finetuning['rougeLsum']['fmeasure']]
}

df = pd.DataFrame(data) # dictionary into a pandas DataFrame

fig, axs = plt.subplots(2, 2, figsize=(14, 12)) # Create a 2x2 grid of subplots

metrics = ['Precision', 'Recall', 'F1-score']  # List of metrics to be plotted
rouge_scores = ['ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'ROUGE-LSUM'] # List of ROUGE score types

for i, score in enumerate(rouge_scores):  # Loop over each ROUGE score type
    row, col = divmod(i, 2)  # row and column index for subplot
    axs[row, col].bar(metrics, df[f'{score} Before'], width=0.4, label='Before Fine-Tuning', align='center')  # Plot bars for scores before fine-tuning
    axs[row, col].bar(metrics, df[f'{score} After'], width=0.4, label='After Fine-Tuning', align='edge')  # Plot bars for scores after fine-tuning
    axs[row, col].set_title(score)  # Set the title of the subplot
    axs[row, col].set_xlabel('Metrics')  # Set the x-axis label
    axs[row, col].set_ylabel('Scores')  # Set the y-axis label
    axs[row, col].legend()  # Add a legend to the subplot

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plot
```
